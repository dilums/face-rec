```python
import numpy as np
import cv2
import face_recognition
import matplotlib.pyplot as plt
import utilities
%matplotlib inline
```

```python
image = face_recognition.load_image_file("images/dicaprio-oscar.jpg")
utilities.show(image,title='Original Image', figsize=(8,8))
```

```python
face_locations = face_recognition.face_locations(image)
```

```python
type(face_locations)
```

```python
print('Number of faces detected : {}'.format(len(face_locations)))
```

```python
top, right, bottom, left = face_locations[0]
print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))
```

```python
utilities.show(image[top:bottom, left:right], title='Detected Face', figsize=(5,5))
```

```python
img_cpy = np.copy(image)
_ = cv2.rectangle(img_cpy,(left,top),(right,bottom),(0,255,0),2)
utilities.show(img_cpy, title='Detected Face Location', figsize=(8,8))
```

```python
face_locations = face_recognition.face_locations(image, number_of_times_to_upsample=0, model="cnn")
print('Number of faces detected : {}'.format(len(face_locations)))
top, right, bottom, left = face_locations[0]
img_cpy = np.copy(image)
_ = cv2.rectangle(img_cpy,(left,top),(right,bottom),(0,255,0),2)
utilities.show(img_cpy,title='Face detected by cnn', figsize=(8,8))
```

```python
image = face_recognition.load_image_file("images/oscar-group.jpg")
face_locations = face_recognition.face_locations(image)
print('Number of faces detected : {}'.format(len(face_locations)))
for face_location in face_locations:
    top, right, bottom, left = face_location
    _ = cv2.rectangle(image,(left,top),(right,bottom),(0,255,0),3)
utilities.show(image)
```

```python
image = face_recognition.load_image_file("images/oscar-group.jpg")
face_locations = face_recognition.face_locations(image, number_of_times_to_upsample=1, model="cnn")
print('Number of faces detected : {}'.format(len(face_locations)))
for face_location in face_locations:
    top, right, bottom, left = face_location
    _ = cv2.rectangle(image,(left,top),(right,bottom),(0,255,0),3)
utilities.show(image)
```
